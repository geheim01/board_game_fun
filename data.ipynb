```python
import pandas as pd
import requests
```

```python
CLIENT_ID = "JLBr5npPhV"
DATA_PATH = "board_games.csv"

MECHANICS_URL = "https://api.boardgameatlas.com/api/game/mechanics?client_id=JLBr5npPhV"
CATEGORIES_URL = "https://api.boardgameatlas.com/api/game/categories?client_id=JLBr5npPhV"
```

```python
df = pd.read_csv(DATA_PATH, sep=";", usecols=range(0,1), names=["name"])
```

```python
def search(game_name:str):
    api_url = f"https://api.boardgameatlas.com/api/search?name={game_name}&client_id={CLIENT_ID}"
    response = requests.get(api_url)
    return response.json()
```

```python
df_mechanics = pd.DataFrame(requests.get(MECHANICS_URL).json()['mechanics'])
df_categories = pd.DataFrame(requests.get(CATEGORIES_URL).json()['categories'])
```

```python
import numpy as np

# Loop through each row and call the search() function for the game name
for index, row in df.iterrows():
    game_name = row['name']
    search_response = search(game_name)

    try:
        game_info = search_response['games'][0]

            # Extract the relevant information from the JSON object and add it to the DataFrame
        df.at[index, 'description'] = game_info.get('description', np.nan)
        df.at[index, 'min_players'] = game_info.get('min_players', np.nan)
        df.at[index, 'max_players'] = game_info.get('max_players', np.nan)
        df.at[index, 'min_age'] = game_info.get('min_age', np.nan)
        df.at[index, 'image_url'] = game_info.get('image_url', np.nan)
        df.at[index, 'designer'] = game_info.get('designer', np.nan)
        df.at[index, 'publisher'] = game_info.get('publisher', np.nan)
        df.at[index, 'artist'] = game_info.get('artist', np.nan)
        df.at[index, 'mechanics'] = ", ".join([df_mechanics[df_mechanics['id']==r['id']]['name'].values[0] for r in game_info.get('mechanics', np.nan)])
        df.at[index, 'categories'] = ", ".join([df_categories[df_categories['id']==r['id']]['name'].values[0] for r in game_info.get('categories', np.nan)])
        df.at[index, 'min_playtime'] = game_info.get('min_playtime', np.nan)
        df.at[index, 'max_playtime'] = game_info.get('max_playtime', np.nan)
        df.at[index, 'year_published'] = game_info.get('year_published', np.nan)
    except:
        print("*** ERROR ***")
        print(game_name)
```
